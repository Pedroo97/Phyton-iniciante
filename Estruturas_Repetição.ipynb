Ler 5 notas e informar a média usando estruturas de repetições while e for


## For

```
nota = media = soma = 0
```

```
print(nota, media, soma)
```

```
for _ in range(1, 6): #range contém os parâmetros start, stop e step, que permitem configurar o retorno do intervalo de diferentes maneiras, como valores múltiplos de um determinado número e até intervalos com números negativos e positivos.
  nota = float(input('Digite a nota '))
  soma += nota
```

```
print(soma)
```

```
media = soma / 5 #dividindo a soma pela quantidade de notas
print('A média é ', media)
```

## While

```
nota = soma = 0
numero = 1
while numero <= 5:
  nota = float(input('Digite a nota:'))
  soma += nota
  numero += 1
print('A média é ', soma / 5)
```
