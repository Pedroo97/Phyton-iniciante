# Dicionários

```python
 coleta = {'Aedes aegypt': 32, 'Aedes albopictus': 22, 'Anopheles darlingi': 14}
```

```python
coleta['Aedes aegypt']
```

```python
coleta['Rhodnius montenegrensis'] = 11
```

```python
print(coleta)
```

```python
del(coleta)['Anopheles darlingi']
print(coleta)
```

```python
coleta.items()
```

```python
coleta.keys()
```

```python
coleta.values()
```

```python
coleta2 = {'Anopheles gambie': 13, 'Anopheles deaneorum': 14}
print(coleta2)
```

```python
coleta.update(coleta2)
print(coleta)
```

```python
for especie, num_especimes in coleta.items():
  print(f'Espécie: {especie}, número de espécimes coletados: {num_especimes}')
```

# Conjuntos (set)

```python
biomoleculas = ('proteína', 'ácidos nucleicos', 'carboidrato', 'lipideo', 'ácidos nucleicos', 'carboidrato', 'carboidrato', 'carboidrato', 'carboidrato')
```

```python
print(biomoleculas)
```

```python
print(set(biomoleculas))
```

```python
c1 = {1,2,3,4,5}
c2 = {3,4,5,6,7}
c3 = c1.intersection(c2)
```

```python
print(c3)
```

```python
c1.difference(c2)
```

```python
c2.difference(c1)
```
