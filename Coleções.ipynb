Crie uma estrutura de repetição para fazer a leitura de 5 números inteiros e os armazene dentro de uma lista. Após a leitura, crie outra estrutura de repetição para somar todos os valores digitados

```
lista = [] #criando uma lista vazia usando colchetes vazios
for _ in range(1, 6): #irá repetir o processo 5 vezes
  valor = int(input('Digite o valor: ')) #criando variável que irá receber o valor que o usuário digitar em formato inteiro
  lista.append(valor)
```

```
lista
```

```
len(lista)
```

```
soma = 0 #criando uma variável
for i in range(len(lista)): #len irá indicar o tamanho da lista
  #print(lista[i])
  soma += lista[i] #operação de soma 
print('Soma: ', soma)
```



Crie um dicionário para armazenar o nome e a nota de 3 alunos, fazendo a leitura dos valores por meio de uma estrutura de repetição. Depois, crie uma nova estrutura de repetição para somar todas as notas e retornar a média

```
alunos = {} #chaves usadas para criar variáveis em dicionários
for _ in range(1, 4):
  nome = input('Digite o nome: ') #para o usuário inserir o nome 
  nota = float(input('Digite a nota: ')) #para o usuário inserir a nota
  alunos[nome] = nota 
```

```
alunos
```

```
alunos['Maria']
```

```
alunos.values() #função usada para ter acesso as notas
```

```
soma = 0
for nota in alunos.values():
  #print(nota)
  soma += nota
print('Média: ', soma / 3)
#operação para se ter o valor da média
```
